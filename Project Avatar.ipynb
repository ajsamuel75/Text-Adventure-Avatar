```python
## Project Avatar Game

import random # using random methods in the game

# stage 1: Arrival on Pandora
def level_1():
  
    #Start image
    print(f"""\n
     █   █ ██▀ █   ▄▀▀ ▄▀▄ █▄ ▄█ ██▀   ▀█▀ ▄▀▄   █▀▄ █▀▄ ▄▀▄   █ ██▀ ▄▀▀ ▀█▀   ▄▀▄ █ █ ▄▀▄ ▀█▀ ▄▀▄ █▀▄
     ▀▄▀▄▀ █▄▄ █▄▄ ▀▄▄ ▀▄▀ █ ▀ █ █▄▄    █  ▀▄▀   █▀  █▀▄ ▀▄▀ ▀▄█ █▄▄ ▀▄▄  █    █▀█ ▀▄▀ █▀█  █  █▀█ █▀▄
    \n""")
    
    #Start Button
    # This block of code is used to validate the player's input 
    # and ensure that they press the enter key to start the game
    introduction = input(prompt = "Hi there!!!\nPlease press Enter to Play.\n")
    while len(introduction) != 0:
        # printing a warning when anything other than the enter key is pressed
        print("INVALID INPUT: PLEASE PRESS ENTER KEY ONLY\n")
        introduction = input(prompt = "Please press Enter to Play\n")
    
    # continuing to the level when enter is pressed
    if len(introduction) == 0:
            pass
    # Introduction
    # This section of the code is used to give the player an overview of the game and their goal
    print(f"Welcome to Project Avatar, a text adventure game inspired by the movie, 'Avatar: The Way of Water'.")
    print("In this very intriguing game, you will take on the role of a human scientist sent to Pandora,\n(a planet far from earth, approx. 4.37 light years in distance) to study the moon's native inhabitants, the Na'vi.")
    # level goal to achieve 
    print("Your goal is to learn as much as you can about the Na'vi and their culture,\nwhile also trying to gain their trust and acceptance.")
    print("But be careful, as there are many dangers on Pandora that you must avoid in order to succeed.")
    
    # level 1 challenge
    print("\nLevel 1: Arrival on Pandora")
    print("You have just arrived on Pandora and are greeted by a group of Na'vi warriors.")
    print("They are suspicious of you and your motives. What do you do?")
    # action choices for the user
    print("1. Introduce yourself and explain your mission.")
    print("2. Remain silent and try to blend in.")
    print("3. Show them the gifts you brought with you as a sign of goodwill.")
    
    # starting overall trust at 0
    trust = 0
    while True:
        # This section of the code is used to get the player's choice and validate their input
        try:
            player_choice = int(input(prompt = "Enter your choice: "))
            if player_choice in range(1,4): # continuing under the condition that 1, 2 or 3 is pressed
                if player_choice == 1:
                    trust_increase = random.randint(1, 5)
                    print("\nYou introduce yourself and explain your mission.\nThe Na'vi warriors seem to be slightly less suspicious.\nThere is less distrust.")
                    trust += trust_increase # increasing trust level randomly if the this choice is taken 
                    break
                elif player_choice == 2: # wrong option for the user to pick
                    trust_decrease = random.randint(1, 5)
                    print("\nYou remain silent and try to blend in. The Na'vi warriors continue to be suspicious.")
                    trust -= trust_decrease # decreasing trust level randomly if the this choice is taken
                    break
                elif player_choice == 3:
                    trust_increase = random.randint(1, 10)
                    print("\nYou show the Na'vi warriors the gifts you brought with you. They seem impressed and your trust with them increases.                     The Na'vi give you permission to take a sample, but emphasize the importance of treating it with respect. Your trust with them increases slightly.")
                    trust += trust_increase # increasing trust level randomly if the this choice is taken
                    break
                else: # condition given if none of the choices are taken
                    print("Invalid choice. Please enter a valid number between 1 and 3.")
            else: # condition given if no number is chosen
                print("Invalid input. Please enter a valid number.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    # returning the 'trust' variable after it has been executed
    return trust

# getting the value(output) returned by the level function and storing it in the variable trust.
trust = level_1()

# stage 2: Learning about the Na'vi
def level_2():
    
    # New level button
    new_level = input(prompt = "Press Enter key to continue to next level.\n")
    
    while len(new_level) != 0:
        # printing a warning when anything other than the enter key is pressed
        print("INVALID INPUT: PLEASE PRESS ENTER KEY ONLY\n")
        # redirecting back to the user input when wrong input is made
        new_level = input(prompt = "Press Enter key to continue to next level.\n")
    if len(new_level) == 0:
        # continuing to the level when enter is pressed
        pass

    #level 2 challenge
    print("\nLevel 2: Learning about the Na'vi")
    print("You spend the next few days living among the Na'vi and learning about their culture.")
    print("You come across a sacred tree that the Na'vi hold dear. What do you do?")
    
    # choices of actions availalable for the user
    print("1. Take a sample for research.")
    print("2. Respect their beliefs and leave it alone.")
    print("3. Ask for permission to take a sample for research.")
    
    # Refreshing the trust back to 0 for a new level
    trust = 0
    while True:
        # This section of the code is used to get the player's choice and validate their input
        try:
            player_choice = int(input(prompt = "Enter your choice: "))
            if player_choice in range(1,4): # continuing under the condition that 1, 2 or 3 is pressed
                if player_choice == 1:
                    trust_decrease = random.randint(1, 5)
                    print("\nYou take a sample of the sacred tree without permission.\nThe Na'vi become angry and your trust with them decreases.\n")
                    trust -= trust_decrease # decreasing trust level randomly if the this choice is taken 
                    break
                elif player_choice == 2:
                    trust_increase = random.randint(1, 5)
                    print("\nYou respect the Na'vi's beliefs and leave the tree alone. Your trust with them increases.")
                    trust += trust_increase # increasing trust level randomly if the this choice is taken 
                    break
                elif player_choice == 3:
                    # most correct answer
                    trust_increase = random.randint(1, 10)
                    print("\nYou ask for permission to take a sample of the sacred tree for research.")
                    print("\nThe Na'vi give you permission, but emphasize the importance of treating it with respect.")
                    trust += trust_increase # increasing trust level randomly if the this choice is taken 
                    break
                else:
                    # warning for if the conditions are not met, 
                    print("Invalid choice. Please enter a valid number between 1 and 3.")
            else: # condition given if no number is chosen
                print("Invalid input. Please enter a valid number.")
        # warning for when anything other than a number is pressed
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    # returning the 'trust' variable after it has been executed
    return trust

# getting the value(output) returned by the level function and storing it in the variable trust.
trust = level_2()


# stage 3: Navigating Pandora
def level_3():

    # New level button
    new_level = input(prompt = "Press Enter key to continue to next level.\n")
    while len(new_level) != 0:
    # printing a warning when anything other than the enter key is pressed
        print("INVALID INPUT: PLEASE PRESS ENTER KEY ONLY\n")
         
        new_level = input(prompt = "Press Enter key to continue to next level.\n")
    if len(new_level) == 0:
    # continuing to the level when enter is pressed
        pass

    #level 3 challenge
    # The title and description of the level
    print("\nLevel 3: Navigating Pandora")
    print(f"""You have to navigate through the different terrains of Pandora to reach your destination.
You have to choose the best path to reach your destination. The terrain is rocky and difficult.""")
    # Path choices available to the user
    print("1. Take the path through the mountains.")
    print("2. Take the path through the jungle.")
    print("3. Take the path through the river.")
    
    # initializing variables
    # Refreshing the trust back to 0 for a new level
    trust = 0
    # creating a variable for path choice
    path_choice = []
    # allowing the player to choose a path
    for path in range(1):
        #using try-except condition for the path options
        try:
            player_choice = int(input(prompt = "Enter your choice: "))
            if player_choice in range(1,4):
            # if the player chooses a valid path, add it to the path_choice list
                if player_choice == 1:
                    path_choice.append("mountains")
                    print(f"\nRocky, your path through Pandora is the", path_choice[0])
                # second choice
                elif player_choice == 2:
                    path_choice.append("jungle")
                    print(f"\nTime for some time in the wild, your path through Pandora is the", path_choice[0])
                # third choice
                elif player_choice == 3:
                    path_choice.append("river")
                    print(f"\nAhoy, your path through Pandora is the", path_choice[0])
                # if the player chooses an invalid option, add to trust variable
                else:
                    print("Invalid choice. Please enter a valid number between 1 and 3.")
                    trust += 5
            # if the player enters an invalid number, prompt them to enter a valid one    
            else:
                print("Invalid input. Please enter a valid number.")
        except ValueError: 
        # if the player enters something other than a number, prompt them to enter a valid one
            print("Invalid input. Please enter a valid number.")
            
    # returning the 'trust' variable after it has been executed
    return trust

# getting the value(output) returned by the level function and storing it in the variable trust.
trust = level_3()

# stage 4: Completing the mission
def level_4():
        
    # New level button
    # Prompting the player to press enter to continue
    new_level = input(prompt = "Press Enter key to continue to next level.\n")
    # validating player input, only allowing the enter key to be pressed
    while len(new_level) != 0:
    # Refreshing the trust back to 0 for a new level
        print("INVALID INPUT: PLEASE PRESS ENTER KEY ONLY\n")
        new_level = input(prompt = "Press Enter key to continue to next level.\n")
    if len(new_level) == 0:
        # continuing to the level when enter is pressed
        pass

    #level 4 challenge
    print("\nLevel 4: Completing the mission")
    print("You have explored Pandora and learned a lot about the Na'vi and their culture.")
    print("It's time to complete your mission and report back to Earth.")
    print("You come across a group of hostile animals. What do you do?")
    
    #  choices available to the user
    print("1. Try to fight them off.")
    print("2. Use your knowledge of the Na'vi to communicate with the animals and peacefully resolve the situation.")
    print("3. Run away.")
    
    # initializing trust variable
    trust = 0
    # running a while loop to get player's choice and validate input
    while True:
        # This section of the code is used to get the player's choice and validate their input
        try:
            player_choice = int(input(prompt = "Enter your choice: "))
            if player_choice in range(1,4):
                # if the player chooses a valid path, add it to the path_choice list
                if player_choice == 1:
                    trust_decrease = random.randint(1, 5)
                    print("\nYou try to fight off the animals, but you're outnumbered and outmatched.\nYour trust with the Na'vi decreases.")
                    trust -= trust_decrease # decreasing trust level randomly if the this choice is taken
                    break
                # second choice    
                elif player_choice == 2:
                    trust_increase = random.randint(1, 10)
                    print("\nYou use your knowledge of the Na'vi to communicate with the animals and peacefully resolve the situation.\nYour trust with the Na'vi increases.")
                    trust += trust_increase # increasing trust level randomly if the this choice is taken
                    break
                # third choice    
                elif player_choice == 3:
                    print("\nYou run away from the animals.\nYour trust with the Na'vi remains unchanged.")
                    break
                else: # condition given if no number is chosen
                    print("Invalid choice. Please enter a valid number between 1 and 3.")
            else:
                # if the player enters an invalid choice, prompt them to enter a valid one
                print("Invalid choice. Please enter a valid number between 1 and 3.")
        except ValueError:
            # if the player enters something other than a number, prompt them to enter a valid one
            print("Invalid input. Please enter a valid number.")
    # returning the 'trust' variable after it has been executed        
    return trust

# getting the value(output) returned by the level function and storing it in the variable trust.
trust = level_4()


# stage 5: Return to Earth
def level_5():
    
    # Riddle game to pass level 4 into final level
    print("\nLet's play a RIDDLE!!!")
    # validating player input, only allowing the enter key to be pressed
    riddle_entrance = input(prompt = "Press Enter key to continue.\n")
    
    # Checking if the user is pressing the Enter key only to continue to the game
    while len(riddle_entrance) != 0:
    # Refreshing the trust back to 0 for a new level    
        print("INVALID INPUT: PLEASE PRESS ENTER KEY ONLY\n")
        riddle_entrance = input(prompt = "Press Enter key to continue.\n")

    if len(riddle_entrance) == 0:
        # continuing to the level when enter is pressed
        pass
    
    # Displaying a riddle for the user to solve to advance to the final level
    print("\nTo advance to the final level, here is an interesting brain teaser riddle given by the Na'vi:\n")
    
    # riddle question
    riddle_input = input(prompt = "I am not alive, but I grow; I don't have lungs, but I need air; I don't have a mouth, but water kills me.\
    \nWhat am I?\n(Fun Fact: The Na'vi people hate it because it destroys)\nAnswer:\n").lower()
    
    # Checking if the user has inputted the correct answer to the riddle
    while "fire" not in riddle_input:
        print(f"""\nINCORRECT ANSWER.\n HINT: 'fir_'\n""")
        print(f"\nLet's try again :)\n")
        # redirecting the user when a wrong answer has been inputted
        riddle_input = input(prompt = "I am not alive, but I grow; I don't have lungs, but I need air; I don't have a mouth, but water kills me.\
        \nWhat am I?\n(Fun Fact: The Na'vi people hate it because it destroys)\nAnswer:\n").lower()
    if "fire" in riddle_input:
        # confirming correct answer
        print("Correct Answer, You are Awesome!!!")
        pass    

    # Final level button
    new_level = input(prompt = "Press Enter key to continue to final level ;) \n")
    
    # Checking if the user is pressing the Enter key only to continue to the final level
    while len(new_level) != 0:
        # validating player input, only allowing the enter key to be pressed
        print("INVALID INPUT: PLEASE PRESS ENTER KEY ONLY\n")
        new_level = input(prompt = "Press Enter key to continue to final level ;) \n")
    if len(new_level) == 0:
        # continuing to the level when enter is pressed
        pass
    
    #Last level challenge
    print("\nLevel 5: Return to Earth")
    print("You have successfully completed your mission and are returning to Earth.")
    print("You must choose how to present your findings to the scientific community.")
    
    # options available to the user
    print("1. Emphasize the potential for exploitation of Pandora's resources.")
    print("2. Highlight the importance of respecting the Na'vi's culture and environment.")
    print("3. Present a balanced view of the potential benefits and risks associated with further study of Pandora.")

    while True:
        # This section of the code is used to get the player's choice and validate their input
        try:
            player_choice = int(input(prompt = "Enter your choice: "))
            if player_choice in range(1,4):
                # if the player chooses a valid path, add it to the path_choice list
                if player_choice == 1:
                    print("\nYou emphasize the potential for exploitation of Pandora's resources. The scientific community is divided on the ethics of your findings.")
                    break
                # second choice    
                elif player_choice == 2:
                    print("\nYou highlight the importance of respecting the Na'vi's culture and environment. Your findings are well-received by the scientific community and the public.")
                    break
                # third choice    
                elif player_choice == 3:
                    print("\nYou present a balanced view of the potential benefits and risks associated with further study of Pandora. Your findings are well-received by the scientific community.")
                    break
                else: # condition given if no number is chosen
                    print("Invalid choice. Please enter a valid number between 1 and 3.")
            else:
                # if the player enters something other than a number, prompt them to enter a valid one
                print("\nInvalid choice. Please enter a valid number between 1 and 3.")
        except ValueError:
            print("\nInvalid input. Please enter a valid number.")
    
level_5()

# conditional option to restart the game
def play_again():
    
    # conditions for the new user-defined function
    while True:
        play_again = input(prompt = "Would you like to play again? (yes/no)\n ").lower()
        if play_again == "yes":
            #start game again
            trust = level_1() # resetting trust level
            level_2()
            level_3()
            level_4()
            level_5()
        elif play_again == "no": #if the user declines to restart the game, an output is set
            # Endgame
            print("\nThanks for playing Avatar: The Way of Water. We hope you enjoyed the adventure and learned something new!")
            break
        else:
            # redirecting the user due to invalid input
            print("\nInvalid input. Please enter 'yes' or 'no'.")
            play_again
def win():
    # user defined function for winning the game
    print("Congratulations! You have successfully gained the trust and acceptance of the Na'vi, explored Pandora and completed the mission.")
    print(f"""
    ┏━━━┓━━━━━━━━━━━━━━━━━━━━━┏┓━━━━━┏┓━━━━━━━┏┓━━━━━━━━━━━━━━━┏┓┏┓┏┓
    ┃┏━┓┃━━━━━━━━━━━━━━━━━━━━┏┛┗┓━━━━┃┃━━━━━━┏┛┗┓━━━━━━━━━━━━━━┃┃┃┃┃┃
    ┃┃━┗┛┏━━┓┏━┓━┏━━┓┏━┓┏━━┓━┗┓┏┛┏┓┏┓┃┃━┏━━┓━┗┓┏┛┏┓┏━━┓┏━┓━┏━━┓┃┃┃┃┃┃
    ┃┃━┏┓┃┏┓┃┃┏┓┓┃┏┓┃┃┏┛┗━┓┃━━┃┃━┃┃┃┃┃┃━┗━┓┃━━┃┃━┣┫┃┏┓┃┃┏┓┓┃━━┫┗┛┗┛┗┛
    ┃┗━┛┃┃┗┛┃┃┃┃┃┃┗┛┃┃┃━┃┗┛┗┓━┃┗┓┃┗┛┃┃┗┓┃┗┛┗┓━┃┗┓┃┃┃┗┛┃┃┃┃┃┣━━┃┏┓┏┓┏┓
    ┗━━━┛┗━━┛┗┛┗┛┗━┓┃┗┛━┗━━━┛━┗━┛┗━━┛┗━┛┗━━━┛━┗━┛┗┛┗━━┛┗┛┗┛┗━━┛┗┛┗┛┗┛
    ━━━━━━━━━━━━━┏━┛┃━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    ━━━━━━━━━━━━━┗━━┛━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    \n""")
def fail():
    # user defined function for losing the game
    print("Game lost! Unfortunately, you were not able to gain the trust and acceptance of the Na'vi and complete your mission.\n")
    print(f"""
    ░▒█▀▀█░█▀▀▄░█▀▄▀█░█▀▀░░░▒█▀▀▀█░▄░░░▄░█▀▀░█▀▀▄
    ░▒█░▄▄░█▄▄█░█░▀░█░█▀▀░░░▒█░░▒█░░█▄█░░█▀▀░█▄▄▀
    ░▒█▄▄▀░▀░░▀░▀░░▒▀░▀▀▀░░░▒█▄▄▄█░░░▀░░░▀▀▀░▀░▀▀
    \n""")

if trust >= 10:
    # conditionally resetting the trust level if game is won
    win()
    play_again()
else:
    # asking the player to play again if game is lost
    fail()
    play_again()

```
